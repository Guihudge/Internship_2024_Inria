```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib widget
```

```python
mS = 1.9e31
mT = 5.9e24
mL = 7.3e22

dTL = 3.8e8
dTS = 1.4e11
dSL = 1.5e9

G = 6.67e-11

time_max = 500
delta_t = 3600 # en seconde (=1h)
```

```python
xS = 0
yS = 0
xT = dTS
yT = 0
xL = dTS+dTL
yL = 0
y = np.array([mT, mS, mL])

vTx = 100
vTy = 100

vSx = 0
vSy = 0

vLx = 120
vLy = 120
```

```python
Ax = np.array([
    [0,G*(mT/dTS**3)*(xT-xS),G*(mT/dTL**3)*(xT-xS)],
    [G*(mS/dTS**3)*(xS-xT),0,G*(mS/dSL**3)*(xS-xL)],
    [G*(mL/dTL**3)*(xL-xT),G*(mL/dSL**3)*(xL-xS),0]
])

Ay = np.array([
    [0,G*(mT/dTS**3)*(yT-yS),G*(mT/dTL**3)*(yT-yS)],
    [G*(mS/dTS**3)*(yS-yT),0,G*(mS/dSL**3)*(yS-yL)],
    [G*(mL/dTL**3)*(yL-yT),G*(mL/dSL**3)*(yL-yS),0]
])
```

```python
Fy = Ay@y
Fx = Ax@y
```

```python
FxT, FxS, FxL = Fx
FyT, FyS, FyL = Fy
```

```python
aTx = 1/mT*FxT
aTy = 1/mT*FyT

aSx = 1/mS*FxS
aSy = 1/mS*FyS

aLx = 1/mL*FxL
aLy = 1/mL*FyL
```

```python
positionTerre = []
positionLune = []
positionSoleil = []

#pour le graphique
#x_min, x_max, y_min, y_max = None
```

```python
for i in range(0,time_max):
    Ax = np.array([
        [0,G*(mT/dTS**3)*(xT-xS),G*(mT/dTL**3)*(xT-xS)],
        [G*(mS/dTS**3)*(xS-xT),0,G*(mS/dSL**3)*(xS-xL)],
        [G*(mL/dTL**3)*(xL-xT),G*(mL/dSL**3)*(xL-xS),0]
    ])
    
    Ay = np.array([
        [0,G*(mT/dTS**3)*(yT-yS),G*(mT/dTL**3)*(yT-yS)],
        [G*(mS/dTS**3)*(yS-yT),0,G*(mS/dSL**3)*(yS-yL)],
        [G*(mL/dTL**3)*(yL-yT),G*(mL/dSL**3)*(yL-yS),0]
    ])

    Fy = Ay@y
    Fx = Ax@y

    FxT, FxS, FxL = Fx
    FyT, FyS, FyL = Fy

    # calcule accélération
    aTx = 1/mT*FxT
    aTy = 1/mT*FyT
    
    aSx = 1/mS*FxS
    aSy = 1/mS*FyS
    
    aLx = 1/mL*FxL
    aLy = 1/mL*FyL

    # calcule vitesse
    vTx = delta_t*aTx + vTx
    vTy = delta_t*aTy + vTy

    vSx = delta_t*aSx + vSx
    vSy = delta_t*aSy + vSy

    vLx = delta_t*aLx + vLx
    vLy = delta_t*aLy + vLy

    #calcule position
    xT = delta_t*vTx + xT
    yT = delta_t*vTy + yT

    xS = delta_t*vSx + xS
    yS = delta_t*vSy + yS

    xL = delta_t*vLx + xL
    yL = delta_t*vLy + yL

    positionTerre.append((xT, yT))
    positionSoleil.append((xS, yS))
    positionLune.append((xL, yL))

    #x_min = min(x_min, min(xT, xL, xS))
    #y_min = min(y_min, min(yT, yL, yS))

    #x_max = max(x_max, max(xT, xL, xS))
    #y_max = max(y_max, max(yT, yL, yS))
```

```python
fig, ax = plt.subplots()
ax.set_xlim(-1.5e11, 1.5e11)  # Arbitrary units for visualization
ax.set_ylim(-1.5e11, 1.5e11)
ax.set_aspect('equal')

sun, = ax.plot([], [], 'yo', markersize=12, label='Sun')
earth, = ax.plot([], [], 'bo', markersize=6, label='Earth')
moon, = ax.plot([], [], 'go', markersize=4, label='Moon')


# Initialize the animation
def init():
    sun.set_data([], [])
    earth.set_data([], [])
    moon.set_data([], [])
    return sun, earth, moon


# Update the animation
def update(frame):
    sunX, sunY = positionSoleil[frame]
    sun.set_data(sunX, sunY)
    return sun, earth, moon


ani = FuncAnimation(fig, update, frames=time_max,
                    init_func=init, blit=True)

# Display the animation
plt.legend()
plt.show()
```
